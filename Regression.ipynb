```python
import pandas as pd
cancer_df = pd.read_csv('data_194228.csv')
air_df = pd.read_csv('Air Quality Data.csv')
cancer_df.head()
```

```python
air_df.head()
```

```python
print(cancer_df.shape[0])
print(air_df.shape[0])
```

```python
# Drop empty columns
air_df_drop = air_df.drop(columns=['Data Comment','Unnamed: 7'])
air_df_drop.head()
```

```python
# Drop empty columns
cancer_df_drop = cancer_df.drop(columns=['Data Comment','Unnamed: 10'])
cancer_df_drop.head()
```

```python
# Split data sets by year
air_2018 = air_df_drop[air_df_drop['Year'] == 2018]
air_2019 = air_df_drop[air_df_drop['Year'] == 2019]
cancer_2018 = cancer_df_drop[cancer_df_drop['Year'] == 2018]
cancer_2019 = cancer_df_drop[cancer_df_drop['Year'] == 2019]
```

```python
# Split 2018 Air Quality data by pollutant
benzene_2018 = air_2018[air_2018['Pollutant'] == 'Pollutant: Benzene']
formaldehyde_2018 = air_2018[air_2018['Pollutant'] == 'Pollutant: Formaldehyde']
acetaldehyde_2018 = air_2018[air_2018['Pollutant'] == 'Pollutant: Acetaldehyde']
tetrachloride_2018 = air_2018[air_2018['Pollutant'] == 'Pollutant: Carbon tetrachloride']
butadiene_2018 = air_2018[air_2018['Pollutant'] == 'Pollutant: 1,3-butadiene']
napthalene_2018 = air_2018[air_2018['Pollutant'] == 'Pollutant: Napthalene']
ethylene_2018 = air_2018[air_2018['Pollutant'] == 'Pollutant: Ethylene oxide']
diesel_2018 = air_2018[air_2018['Pollutant'] == 'Pollutant: Diesel Particulate Matter']
```

```python
# Split 2019 Air Quality data by pollutant
benzene_2019 = air_2019[air_2019['Pollutant'] == 'Pollutant: Benzene']
formaldehyde_2019 = air_2019[air_2019['Pollutant'] == 'Pollutant: Formaldehyde']
acetaldehyde_2019 = air_2019[air_2019['Pollutant'] == 'Pollutant: Acetaldehyde']
tetrachloride_2019 = air_2019[air_2019['Pollutant'] == 'Pollutant: Carbon tetrachloride']
butadiene_2019 = air_2019[air_2019['Pollutant'] == 'Pollutant: 1,3-butadiene']
napthalene_2019 = air_2019[air_2019['Pollutant'] == 'Pollutant: Napthalene']
ethylene_2019 = air_2019[air_2019['Pollutant'] == 'Pollutant: Ethylene oxide']
diesel_2019 = air_2019[air_2019['Pollutant'] == 'Pollutant: Diesel Particulate Matter']
```

# Linear Regression - 2018

Scatter plots of pollutants and cancer in 2018

```python
import sklearn
from sklearn.linear_model import LinearRegression
cancer_2018_float = cancer_2018['Value'].str.strip('%').astype(float) / 100
# Linear bt 2018 Benzene and Cancer
model1 = LinearRegression()
model1.fit(benzene_2018[['Value']], cancer_2018_float)
score1 = model1.score(benzene_2018[['Value']], cancer_2018_float)
print(f'Benzene: {score1}') # R-squared
benz_pred = model1.predict(benzene_2018[['Value']])
# Formaldehyde
model2 = LinearRegression()
model2.fit(formaldehyde_2018[['Value']], cancer_2018_float)
print('Formaldehyde: ' + str(model2.score(formaldehyde_2018[['Value']], cancer_2018_float)))
form_pred = model2.predict(formaldehyde_2018[['Value']])
# Acetaldehyde
model3 = LinearRegression()
model3.fit(acetaldehyde_2018[['Value']], cancer_2018_float)
print('Acetaldehyde: ' + str(model3.score(acetaldehyde_2018[['Value']], cancer_2018_float)))
acet_pred = model3.predict(acetaldehyde_2018[['Value']])
# Tetrachloride
model4 = LinearRegression()
model4.fit(tetrachloride_2018[['Value']], cancer_2018_float)
print('Carbon Tetrachlorodione: ' + str(model4.score(tetrachloride_2018[['Value']], cancer_2018_float)))
tet_pred = model4.predict(tetrachloride_2018[['Value']])
# Butadiene
model5 = LinearRegression()
model5.fit(butadiene_2018[['Value']], cancer_2018_float)
print('Butadiene: ' + str(model5.score(butadiene_2018[['Value']], cancer_2018_float)))
bu_pred = model5.predict(butadiene_2018[['Value']])
# Napthalene
model6 = LinearRegression()
model6.fit(napthalene_2018[['Value']], cancer_2018_float)
print('Napthalene: ' + str(model6.score(napthalene_2018[['Value']], cancer_2018_float)))
nap_pred = model6.predict(napthalene_2018[['Value']])
# Diesel
model7 = LinearRegression()
model7.fit(diesel_2018[['Value']], cancer_2018_float)
print('Diesel Particulate Matter: ' + str(model7.score(diesel_2018[['Value']], cancer_2018_float)))
diesel_pred = model7.predict(diesel_2018[['Value']])
```
